```python
# QUESTION 2
# Assume a and b are two (20, 20) numpy arrays. 
# The L2-distance (defined above) between two equal dimension arrays can be calculated in python as follows:

import numpy as np


def l2_dist(a, b):
     result = ((a - b) * (a - b)).sum()
     result = result ** 0.5
     return result

a = np.random.randint(10, size=(20, 20))
b = np.random.randint(10, size=(20, 20))

try:
    print(np.reshape(b, (20 * 20), 1))
except:
    print("This will give an error on l2_dist!")
    
try:
    print(l2_dist(np.reshape(a, (20 * 20)), np.reshape(b, (20 * 20), 1)))
except Exception as error:
    print(f"See?: {error}")
    
print(l2_dist(a, b))
print(l2_dist(np.reshape(a, (20 * 20)), np.reshape(b, (20 * 20))))
print(l2_dist(a.T, b.T))
```

```python
# QUESTION 5
# Given the 6x6 NumPy array r shown below, which of the following options would slice the shaded elements?

r = np.linspace(0, 35, 36).reshape(6, 6)

print(r)
print(r[[2,4], [2, 4]]) # incorrect: return elements at [2, 2] and [4, 4]
print(r[2:4, 2:4])      # correct
print(r[2:3, 2:3])      # incorrect: wrong end index
print(r[[2,3], [2, 3]]) # incorrect: return elements at [2, 2] and [3, 3]
```

```python
# QUESTION 6
# For the given string, which of the following regular expressions
# can be used to check if the string starts with 'AC'?

import re


s = 'ACBCAC'

print(re.findall('AC', s))      # incorrect: finds all occurences of 'AC'
print(re.findall('[^A]C', s))   # incorrect: finds all occurences that end with 'C' but don't start with 'A'
print(re.findall('^[AC]', s))   # incorrect: finds all occurences at the beginning of the string that are 'A' or 'C'
print(re.findall('^AC', s))     # correct

# more correct
print(re.search('^AC', s))
```

```python
# QUESTION 7
# What will be the output of the variable L after the following code is executed?

import re


s = 'ACAABAACAAAB'
result = re.findall('A{1,2}', s)  # finds all occurences of 'A' and 'AA': *A*_C_*AA*_B_*AA*_C_*AA*_*A*_B -> 5

L = len(result)
print(L)
```

```python
# QUESTION 8
# Which of the following is the correct regular expression to extract all the 
# phone numbers from the following chunk of text:

chunk="""
Office of Research Administration: (734) 647-6333 | 4325 North Quad
Office of Budget and Financial Administration: (734) 647-8044 | 309 Maynard, Suite 205
Health Informatics Program: (734) 763-2285 | 333 Maynard, Suite 500
Office of the Dean: (734) 647-3576 | 4322 North Quad
UMSI Engagement Center: (734) 763-1251 | 777 North University
Faculty Adminstrative Support Staff: (734) 764-9376 | 4322 North Quad
"""

my_pattern = '[(]\d{3}[)]\s\d{3}[-]\d{4}'

print(re.findall(my_pattern, chunk))

answers = {
    "answer 1": '\d{3}[-]\d{3}[-]\d{4}',
    "answer 2": '\d{3}\s\d{3}[-]\d{4}',
    "answer 3": '[(]\d{3}[)][-]\d{3}[-]\d{4}',
    "answer 4": '[(]\d{3}[)]\s\d{3}[-]\d{4}',
              }

for answer, pattern in answers.items():
    print(f"{answer}:, {re.findall(pattern, chunk)}")


    
```

```python
# QUESTION 9
# Which of the following regular expressions can be used to get the domain names (e.g. google.com, www.baidu.com)
# from the following sentence?

text = 'I refer to https://google.com and I never refer http://www.baidu.com if I have to search anything'

my_pattern = '[\w]*\.[\w|.]*'

print(re.findall(my_pattern, text))

answers = {
    "answer 1": '(?<=[https]:\/\/)([A-Za-z0-9.]*)',
    "answer 2": '(?<=https:\/\/)([A-Za-z0-9]*)',
    "answer 3": '(?<=https:\/\/)([A-Za-z0-9.]*)',
    "answer 4": '(?<=[https]:\/\/)([.]*)',
              }

for answer, pattern in answers.items():
    print(f"{answer}:, {re.findall(pattern, text)}")
```

```python
# QUESTION 10
# The text from the Canadian Charter of Rights and Freedoms section 2 lists the fundamental freedoms
# afforded to everyone. Of the four choices provided to replace X in the code below,
# which would accurately count the number of fundamental freedoms that Canadians have?

text=r'''Everyone has the following fundamental freedoms:
(a) freedom of conscience and religion;
(b) freedom of thought, belief, opinion and expression, including freedom of the press and other media of communication;
(c) freedom of peaceful assembly; and
(d) freedom of association.
'''

import re

# pattern = X
my_pattern = '\('           # find all occurrences of '('

print(len(re.findall(my_pattern, text)))

answers = {
    "answer 1": '\(.\)',    # find all occurrences of '(', followed by any other character ('.' is a wildcard), followed by a ')'
    "answer 2": '[a-d]',    # find all occurrences of letters 'a', 'b', 'c' and 'd'
    "answer 3": 'freedom',  # find all occurrences of the string 'freedom'
    "answer 4": '(.)',      # find all occurences of any character except newline
              }

for answer, pattern in answers.items():
    print(f"{answer}:, {len(re.findall(pattern, text))}")
```
